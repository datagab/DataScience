```python
# Crie um banco de dados no SQLite, crie uma tabela, insira registros, 
# consulte a tabela e retorne os dados em dataframe do Pandas
import sqlite3
con = sqlite3.connect('meubanco.db')
cur = con.cursor()
```

```python
sql_create = 'create table leitura '\
'(id integer primary key, '\
'livros varchar(100), '\
'paginas integer, '\
'categoria varchar(140))'
```

```python
cur.execute(sql_create)
```

```python
sql_insert = 'insert into leitura values (?, ?, ?, ?)'
# Dados
recset = [(1000, 'Livro A', 500, 'Romance'),
          (1001, 'Livro B', 767, 'Tech'),
          (1002, 'Livro C', 32,  'Acao')]
# Inserindo os registros
for rec in recset:
    cur.execute(sql_insert, rec)
```

```python
sql_select = 'select * from leitura'
cur.execute(sql_select)
dados = cur.fetchall()
```

```python
for linha in dados:
    print ('Livro Id: %d, Livro: %s, Pagina: %d Categoria: %s \n' % linha)

    
```

```python
import pandas as pd
df = pd.DataFrame(dados)
df
```
